# spark의 데이터 구조 총 3가지 - RDD, Dataframe, DataSet
- 차이 제대로 잘 모름, 궁금하면 찾아보기를
- RDD - 데이터가 비구조적인 경우 사용하기 적합, spark의 핵심
- Dataframe - 구조적인 경우

# RDD - python list를 sparkcontext.parallelize() 함수를 통해 RDD 생성

```python
myList=[1,2,3,4,5,6,7]
```

```python
myRdd = spark.sparkContext.parallelize(myList)
```

```python
myRdd.take(3)
```

```python
myRdd.collect()
```
