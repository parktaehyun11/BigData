```python
bicycle = spark.read.format('com.databricks.spark.csv').options(header='true', inferschema='true').load('data/bicycle.csv')
#컬럼명 변경
bicycle=bicycle\
.withColumnRenamed("date", "Date")\
.withColumnRenamed("count", "Count")
```

```python
bicycle.show(5)
```

```python
bicycle=bicycle.withColumn("year",bicycle.Date.substr(1, 4)) #첫번째부터 4글자 year column 생성
bicycle=bicycle.withColumn("month",bicycle.Date.substr(6, 2)) # 6번째부터 2글자 month column 생성
```

```python
bicycle.show(5)
#bicycle.take(3)
```

```python
bicycle.groupBy('year').agg({"Count":"sum"}).show()
```

```python
bicycle.groupBy('year').pivot('month').agg({"Count":"sum"}).show()
```
